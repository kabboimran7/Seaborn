# Line plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var = [1,2,3,4,5,6,7]
var_1 = [2,3,4,1,5,6,7]
```

```python
# Using matplotlib
plt.plot(var,var_1)
plt.show()
```

```python
#Using Seaborn  - This works best with dataframe 
'''But we can use raw lists, NumPy arrays, or dictionaries 
in some Seaborn functions, but you'll need to explicitly 
pass values for x and y, and not use the data= argument.'''

df = pd.DataFrame({'var':var, 'var_1':var_1}) 

sns.lineplot(x='var',y='var_1', data=df)

# To remove this line : <Axes: xlabel='var', ylabel='var_1'>
plt.show()
```

```python
#Another appraoch : (Direct var/ parameter pass) . Eta X and Y axix k label kore na

sns.lineplot(x=var,y=var_1)
plt.show()
```

# Using seaborns datasets (seaborn.load_dataset)

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
sns.lineplot(x='bill_length_mm',y='body_mass_g', data=penguins)
plt.show()
```

```python
'''1.🎨 hue — Color grouping

The hue parameter adds color to your plot based on a categorical or numerical variable.
It groups the data and uses different colors for each group.

2. 📏 size — Bubble size / marker size

The size parameter changes the size of each point based on a variable (usually numeric).
This is useful when you want to show a third numeric variable in a 2D plot.'''
```

```python
# using multiple parameters

sns.lineplot(x='bill_length_mm',y='body_mass_g', data=penguins, hue='sex')
plt.show()
```

```python
sns.lineplot(x='bill_length_mm',y='body_mass_g', data=penguins, hue='sex', size =10)
plt.show()
```

```python
sns.lineplot(x='bill_length_mm',y='body_mass_g', data=penguins, hue='sex', size =90)
plt.show()
```

```python
sns.lineplot(x='bill_length_mm',y='body_mass_g', data=penguins, hue='sex', style='sex', palette='Accent')
plt.show()
```

```python
sns.lineplot(x='bill_length_mm',y='body_mass_g', data=penguins, hue='sex', style='sex', markers=['o','>'])
plt.show()
```

```python
penguins = sns.load_dataset('penguins').head(20)
```

```python
sns.lineplot(x='bill_length_mm',y='body_mass_g', data=penguins, hue='sex', 
             style='sex', markers=['o','>'])
plt.show()
```

```python
sns.lineplot(x='bill_length_mm',y='body_mass_g', data=penguins, hue='sex', 
             style='sex', markers=['o','>'], dashes=False)   #Ekhane false deway dot dot chole gese 
plt.show()
```

```python
sns.lineplot(x='bill_length_mm',y='body_mass_g', data=penguins, hue='sex', 
             style='sex', markers=['o','>'], legend='full')  #Ekhane false deway dot dot chole gese 

plt.grid()   #grid add kroe 
plt.title('Python')  #title add kore
plt.show()
#Ekhane matplotlib er sob function use kora jabe.
```
