# Histogram Plot 

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = sns.load_dataset('penguins')
var.head()
```

```python
sns.displot(var['flipper_length_mm'])
plt.show()
```

```python
# bins use kore limit set kora hoy jate kore data fix range er modde ashe
sns.displot(var['flipper_length_mm'],bins=[170,180,190,200,210,220,230,240])
plt.show()
```

```python
sns.displot(var['flipper_length_mm'], kde=True) #Kde gives  a line graph in histogram
plt.show()

'''In statistics, KDE stands for Kernel Density Estimation. It's a non-parametric method used to estimate the
probability density function (PDF) of a random variable. Essentially, it helps visualize the distribution of 
data points by creating a smooth curve that represents the likelihood of values falling within certain ranges, 
similar to a histogram but with a smoother, continuous representation. '''
```

```python
'''A rug plot is a statistical visualization that represents the distribution of
a single variable by displaying small lines (rugs) along an axis, indicating the
presence of data points at those values. It's essentially a histogram with zero-width
bins or a one-dimensional scatter plot. '''

sns.displot(var['flipper_length_mm'], bins=[170,180,190,200,210,220,230,240], kde=True , rug= True, color='g') 
plt.show()
```

```python
'''A log scale, or logarithmic scale, uses a transformed scale to show changes in values, 
particularly when dealing with data that spans a wide range of magnitudes or exhibits 
exponential growth. Instead of representing values directly, it represents them as powers
of a base number, usually 10. This allows for a more compressed representation, making it 
easier to visualize and compare values that differ by orders of magnitude. '''

sns.displot(var['flipper_length_mm'],  kde=True , rug= True, color='g', log_scale = True) 
plt.show()
```

```python
sns.displot(var['bill_length_mm'].head(20), kde=True , rug= True, color='g', log_scale = True) 
plt.show()
```

# Barplot

```python
med = pd.read_csv('Medicaldataset.csv') 
med.head()
```

```python
sns.barplot(x = med['Gender'], y = med['Blood sugar'])
plt.show()
```

```python
sns.barplot(x = 'Gender', y = 'Blood sugar', data=med, hue='Result') # using hue I can differentiate / divide the columns in more pieces for better visualization
plt.show()
```

```python
#If I want to create an order then that is also possible . For instance I want 1 brefore o in X axis
order_1 = [1,0]
sns.barplot(x = 'Gender', y = 'Blood sugar', data=med, hue='Result', order = order_1)
plt.show()
```

```python
#For better understanding Here I am sing another dataset: 
var = sns.load_dataset('penguins')
var.head()
```

```python
sns.barplot(x='island', y='bill_length_mm', data=var, hue='sex')
plt.show()
```

```python
#here I am making a new order of columns than the previous one
order_2 = ['Biscoe','Dream','Torgersen']
sns.barplot(x='island', y='bill_length_mm', data=var, hue='sex', order= order_2)
plt.show()
```

```python
#If I wnat to bring Feale infront of male then I gotta change order of hue parameter
sns.barplot(x='island', y='bill_length_mm', data=var, hue='sex', hue_order=['Female','Male'], order= order_2)
plt.show()
```

```python
''' 
    "ci" controls the display and calculation of confidence intervals, while "n_boot" controls the number of bootstrap 
    samples used to compute the interval when bootstrapping is employed.Together, they provide flexibility in visualizing
    and estimating the uncertainty in statistical results. 
'''

sns.barplot(x='island', y='bill_length_mm', data=var, hue='sex', ci=100, n_boot=100)
plt.show()
```

```python
# If I want to create Horizontal barplot (using orient parameter)hen I have to take numerical columns . Stings /  Categorical datas r not allowed
sns.barplot(x='bill_length_mm', y='body_mass_g', data=var, orient='v')
plt.show()
```

```python
# If I want to create Horizontal barplot (using orient parameter)hen I have to take numerical columns . Stings /  Categorical datas r not allowed
sns.barplot(x='bill_length_mm', y='body_mass_g', data=var, orient='h')
plt.show()
```

```python
sns.barplot(x='bill_length_mm', y='body_mass_g', data=var.head(20), orient='v')
plt.show()
```

```python
sns.barplot(x='island', y='bill_length_mm', data=var, hue='sex', palette='icefire')
plt.show()
```

```python
#Saturation color fade/ deep kore , errcolor diye errror er  color change hoy.
sns.barplot(x='island', y='bill_length_mm', data=var, hue='sex', saturation=1, errcolor ='r', errwidth=12)
plt.show()
```

```python
# capsize -----

'''psize is a parameter used to add or adjust the caps on error bars in plots 
like barplot, pointplot, and lineplot. Error bars visually represent the uncertainty 
or variability in data, and the capsize parameter enhances their readability by adding
horizontal lines (caps) at the ends of the error bars. '''


sns.barplot(x='island', y='bill_length_mm', data=var, hue='sex',capsize=.5 )
plt.show()
```

```python
# dodge (if false then one column overlaps on other)
'''used to separate the dots for different levels of a categorical variable.
By default, dodge is set to False , meaning that dots for different levels are not 
eparated. Python.'''

sns.barplot(x='island', y='bill_length_mm', data=var, hue='sex',dodge=False )
plt.show()
```

```python
# Alpha parameter is used to make the barplot blur
sns.set(style='darkgrid')   # using darkgrid
sns.barplot(x='island', y='bill_length_mm', data=var, hue='sex', saturation=100)
plt.show()
```

```python
sns.set(style='dark') # using dark
sns.barplot(x='island', y='bill_length_mm', data=var, hue='sex', saturation=100)
plt.show()
```

# <b> Style options:

'darkgrid' (default) – Grid with dark background.

'whitegrid' – Grid with white background (good for plots with categorical data).

'dark' – No grid, dark background.

'white' – No grid, white background.

'ticks' – White background with ticks on axes.

# <b>context options (controls font/element sizes):

'paper' – For publication.

'notebook' – Default for Jupyter.

'talk' – Bigger elements for presentations.

'poster' – Even larger for posters.

# Scatter PLot

```python
med = pd.read_csv('Medicaldataset.csv') 
med.head()
```

```python
sns.set_style('white')
sns.set_context('paper')  # poster/notebook/talk etc 
sns.scatterplot(x='Age', y='Blood sugar', data=med.head(100), hue='Result', style='Result', size='Result') # Result parameter use kore Female er size kome gese 
plt.show()
```

```python
# To set size according to my wish I gotta use sizes and the columns will be changed accordingly
sns.scatterplot(x='Age', y='Blood sugar', data=med.head(50), hue='Result', style='Result',size='Result', sizes=(100,10)) # Result parameter use kore Female er size kome gese 
plt.show()
```

```python
m = {'negative':'>', 'positive':'*'}
sns.scatterplot(x='Age', y='Blood sugar', data=med.head(100), hue='Result', 
                style='Result', palette ='gist_ncar', alpha=.4, markers=m)
plt.show()
```
